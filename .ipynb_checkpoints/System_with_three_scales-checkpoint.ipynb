<big> <big> <big> System with three scales

```python
""" Import modules"""

import numpy as np
import matplotlib.pyplot as plt
```

```python
"""Define constants of the system"""

I = complex(0.,1.)

psi = 0.
omega1 = 2.
omega2 = 20.
omega3 = 200.
r1 = omega1*I + psi
r2 = omega2*I + psi
r3 = omega3*I + psi

init_cond = (1. + 0.*I)*np.ones(3,complex)
```

```python
"""Coarsening factor"""

coarsening_factor = int(10)
```

$\Large \text{Define the averaged and unaveraged problems.}$

```python
"""Define averaged and unaveraged right handsides."""

def unaveraged_rhs_3d(v,t):
    """Unaveraged right handside"""
    
    v1 = v[0]; v2 = v[1]; v3 = v[2]
    
    dot_v1 = (-1.)*np.exp((-r1)*t)*v1*v1
    dot_v2 = (-1.)*np.exp((-r2)*t)*v2*v2
    dot_v3 = (-1.)*np.exp((-r3)*t)*v3*v3
    
    dot_v = np.zeros(3, complex)
    dot_v[0] = dot_v1; dot_v[1] = dot_v2; dot_v[2] = dot_v3
    
    return dot_v


def averaged_rhs_3d_level_1(v,t):
    """Averaged right handside, weak averaging."""
    
    v1 = v[0]; v2 = v[1]; v3 = v[2]
    
    # averaging window of length 0.1 
    kappa1 = 0.9996175745253061+1.0359109980530905e-16j
    kappa2 = 0.9623377268354719+1.0568490120641062e-16j
    kappa3 = -0.02943966559463109-9.642177234614599e-16j
    
    dot_v1 = kappa1*(-1.)*np.exp((-r1)*t)*v1*v1
    dot_v2 = kappa2*(-1.)*np.exp((-r2)*t)*v2*v2
    dot_v3 = kappa3*(-1.)*np.exp((-r3)*t)*v3*v3
    
    dot_v = np.zeros(3, complex)
    dot_v[0] = dot_v1; dot_v[1] = dot_v2; dot_v[2] = dot_v3
    
    return dot_v


def averaged_rhs_3d_level_2(v,t):
    """Averaged right handside, strong averaging."""
    
    v1 = v[0]; v2 = v[1]; v3 = v[2]
    
    # averaging window of length 1.0 
    kappa1 = 0.9623377268354685-1.0511106577711771e-16j
    kappa2 = -0.02943966559463252-1.0987700171391534e-15j
    kappa3 = -6.530936811434557e-09-7.297986004140192e-17j
    
    dot_v1 = kappa1*(-1.)*np.exp((-r1)*t)*v1*v1
    dot_v2 = kappa2*(-1.)*np.exp((-r2)*t)*v2*v2
    dot_v3 = kappa3*(-1.)*np.exp((-r3)*t)*v3*v3
    
    dot_v = np.zeros(3, complex)
    dot_v[0] = dot_v1; dot_v[1] = dot_v2; dot_v[2] = dot_v3
    
    return dot_v
```

$\Large \text{Implement Runge-Kutta solvers}$

```python
"""Runge-Kutta solvers"""

def RK4_step(v0,t,f,dt):
    """One step with RK4 method. Can be easily adjusted to a first or second order RK method."""
    
    k1 = f(v0,t)
    k2 = f(v0+0.5*dt*k1, t+0.5*dt)
    k3 = f(v0+0.5*dt*k2, t+0.5*dt)
    k4 = f(v0+dt*k3, t+dt)
    
    v1 = v0 + dt/6. * (k1 + 2.*k2 + 2.*k3 + k4)
    
    return v1


def RK2_step(v0,t,f,dt):
    """One step with a RK2 method."""
    
    k1 = f(v0,t)
    k2 = f(v0+0.5*dt*k1, t+0.5*dt)
    
    v1 = v0 + dt * k2
    
    return v1
```

$\Large \text{Compute a reference solution with the RK4 method.}$

```python
def solution_with_RK4(f, Tmin_RK4, Tmax_RK4, dt_RK4):
    
    grid_RK4 = np.arange(Tmin_RK4, Tmax_RK4+0.1*dt_RK4, dt_RK4)
    L_RK4 = len(grid_RK4)
    
    solution_RK4 = np.zeros((3,L_RK4), complex)
    solution_RK4[:,0] = init_cond 
    
    for n in range(L_RK4-1):
        solution_RK4[:,n+1] = RK4_step(solution_RK4[:,n],grid_RK4[n],f,dt_RK4)
        
    
    return grid_RK4, solution_RK4

# plot
grid_RK4_test, solution_RK4_test = solution_with_RK4(unaveraged_rhs_3d, 0, 6, 0.000025);
level1_grid_test, level1_solution_test = solution_with_RK4(averaged_rhs_3d_level_1, 0, 6, 0.001);
level2_grid_test, level2_solution_test = solution_with_RK4(averaged_rhs_3d_level_2, 0, 6, 0.001);

#plt.plot(grid_RK4_test, solution_RK4_test[0,:].real, linewidth=2)
#plt.plot(level1_grid_test, level1_solution_test[0,:].real,linewidth=2)
#plt.plot(level2_grid_test, level2_solution_test[0,:].real,linewidth=2)
#plt.legend(['no averaging','light averaging', 'strong averaging'],bbox_to_anchor=(1.5, 1.05))
#plt.show()

#plt.xlim(0.,1)
#plt.plot(grid_RK4_test, solution_RK4_test[1,:].real)
#plt.plot(level1_grid_test, level1_solution_test[1,:].real,'--')
#plt.plot(level2_grid_test, level2_solution_test[1,:].real,':')
#plt.show()

#plt.xlim(0.,.1)
#plt.plot(grid_RK4_test, solution_RK4_test[2,:].real)
#plt.plot(level1_grid_test, level1_solution_test[2,:].real,'*')
#plt.plot(level2_grid_test, level2_solution_test[2,:].real,':')
#plt.show()

```


$\\ \Large \text{Apply a 3-level scheme to compute a solution to the problem and investigate the error.}$

```python
def twoLevelParareal_as_finePropagator(Tmin_1, Tmax_1, dt_1, fine_dt_1, ic_1):
    """This function is the fine propagator of the three-level method. It is a two-level Parareal method. 
    The coarse propagator in this function uses the weakly averaged RHS."""
    
    # define the coarse propagator
    coarse_prop = averaged_rhs_3d_level_1
      
    # number of Parareal correction iterations
    numIt_1 = 3   
    
    # define coarse grid 
    grid_1 = np.arange(Tmin_1, Tmax_1+0.1*dt_1, dt_1)
    L_1 = len(grid_1)     
    
    # define fine grid
    fine_grid_1 = np.arange(0, dt_1+0.5*fine_dt_1, fine_dt_1)
    fine_L_1 = len(fine_grid_1) 
    
    # array for all Parareal iteration steps
    solution_1 = np.zeros((numIt_1+1,3,L_1), complex) 
    
    # array for Parareal correction
    level_1_sol = np.zeros((3,L_1), complex)
    level_1_sol[:,0] = ic_1
    
    # array for initial guess
    init_guess_1 = np.zeros((3,L_1), complex)
    init_guess_1[:,0] = ic_1
    
    
    # compute initial guess, serial steps
    for n in range(L_1-1):
        init_guess_1[:,n+1] = RK2_step(init_guess_1[:,n],grid_1[n],coarse_prop,dt_1)
        
    # save initial guess    
    solution_1[0,:,:] = init_guess_1.copy()
    
    
    # do numIt_1 Parareal correction iterations
    for k1 in range(numIt_1):
    
        # Parareal iterations
        
        # initialize array to store solution from the fine propagator at the coarse grid points
        fine_prop_1 = np.zeros((3,L_1), complex)
        fine_prop_1[:,0] = init_guess_1[:,0].copy()
    
        # initialize array for solution from the fine propagator at the fine grid points
        f_prop_1 = np.zeros((3,fine_L_1), complex)
        f_prop_1[:,0] = init_guess_1[:,0].copy()
    
        # loop over coarse grid points, can be done in parallel
        for n in range(L_1-1):
            # initial value
            f_prop_1[:,0] = init_guess_1[:,n].copy()
            
            # compute solutions on the fine grids
            for m in range(fine_L_1-1):
                # fine grid point
                time = grid_1[n] + fine_grid_1[m]  
                # solution at fine grid point
                f_prop_1[:,m+1] = RK2_step(f_prop_1[:,m],time,unaveraged_rhs_3d,fine_dt_1)
                  
            # solution of the fine propagator  
            fine_prop_1[:,n+1] = f_prop_1[:,fine_L_1 -1].copy()         
        
        # loop over coarse grid points, cannot be done in parallel
        for n in range(L_1-1):  
            old_sol_1 = RK2_step(init_guess_1[:,n],grid_1[n],coarse_prop,dt_1)
            new_sol_1 = RK2_step(level_1_sol[:,n],grid_1[n],coarse_prop,dt_1)
            level_1_sol[:,n+1] = new_sol_1 + fine_prop_1[:,n+1].copy() - old_sol_1 
            
        
        # save solution from the current Parareal correction step
        solution_1[k1+1,:,:] = level_1_sol.copy()
        
        # update initial guess 
        init_guess_1 = level_1_sol.copy()
        
        level_1_sol = np.zeros((3,L_1), complex)
        level_1_sol[:,0] = ic_1

  
    return grid_1, init_guess_1

```

```python
"""Implement a three-level Parareal method with averaging."""

def threeLevel_Parareal_solution(Tmin_2, Tmax_2, dt_2, fine_dt_2, ic_2):
    """Three-level Parareal method. The coarse propagator uses the strongly averaged RHS."""
    
    # define the coarse propagator
    coarse_propagator  = averaged_rhs_3d_level_2
       
    # number of Parareal correction iterations
    num_it_2 = 7    
      
    # coarse grid 
    grid_2 = np.arange(Tmin_2, Tmax_2+0.1*dt_2, dt_2)
    L_2 = len(grid_2)   
     
    # fine grid 
    fine_grid_2 = np.arange(0, dt_2+0.1*fine_dt_2, fine_dt_2)
    fine_L_2 = len(fine_grid_2)
    
    
    # array to save all Parareal correction iterations
    solutions_2 = np.zeros((num_it_2+1, 3, L_2),complex)
    
    # array for initial guess
    init_guess_2 = np.zeros((3,L_2), complex)
    init_guess_2[:,0] = ic_2
    
    # array for Parareal correction iterations
    level_2_sol = np.zeros((3,L_2), complex)
    level_2_sol[:,0] = ic_2
    
    
    # compute initial guess, serial steps  
    for n2 in range(L_2-1):
        init_guess_2[:,n2+1] = RK2_step(init_guess_2[:,n2],grid_2[n2],coarse_propagator,dt_2)
    
    # save initial guess
    solutions_2[0,:,:] = init_guess_2.copy()
    
    
    # loop over Parareal correction iterations
    for k in range(num_it_2):
               
        # Two-level Parareal as fine propagator, can be applies in parallel       
        # initialize array for solutions from the fine propagator at the coarse grid points
        fine_prop_2 = np.zeros((3,L_2), complex)
        fine_prop_2[:,0] = init_guess_2[:,0].copy()
    
        # loop over fine time steps, work can be done in parallel
        for n2 in range(L_2-1):
            fine_grid_2, fine_sol_2 = twoLevelParareal_as_finePropagator(grid_2[n2] , grid_2[n2+1],  \
                                        fine_dt_2,  fine_dt_2/(coarsening_factor), init_guess_2[:,n2])
            fine_prop_2[:,n2+1] = fine_sol_2[:,len(fine_grid_2)-1].copy()
            
                                      
        # Parareal correction iteration, must be done in serial
        for n2 in range(L_2-1):
            old_sol_2 = RK2_step(init_guess_2[:,n2],grid_2[n2],coarse_propagator,dt_2)
            new_sol_2 = RK2_step(level_2_sol[:,n2],grid_2[n2],coarse_propagator,dt_2)
            level_2_sol[:,n2+1] = new_sol_2 + fine_prop_2[:,n2+1].copy() - old_sol_2 
            
    
        # save solution from current Parareal iteration
        solutions_2[k+1,:,:] = level_2_sol.copy()
        
        # update initial guess
        init_guess_2 = level_2_sol.copy()
        
        level_2_sol = np.zeros((3,L_2), complex)
        level_2_sol[:,0] = ic_2
        
    
    return grid_2, solutions_2 
   
```
